K-means Clustering:

K-means is a popular unsupervised clustering algorithm that partitions the data into K clusters.
In this method, you would first convert your sentences into numerical vectors using techniques like TF-IDF (Term Frequency-Inverse Document Frequency) or word embeddings (e.g., Word2Vec, GloVe).
Then, you apply the K-means algorithm to cluster these vectors into K groups.
The advantage of K-means is its simplicity and efficiency, making it suitable for large datasets.
In this method, you specify the number of clusters K.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans


# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()
# Transform the claims into a TF-IDF matrix
X = vectorizer.fit_transform(claims)

k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k) # Initialize the KMeans clustering algorithm with the specified number of clusters
kmeans.fit(X) # Fit the KMeans model to the TF-IDF matrix

# Assign each sentence to its corresponding cluster
clusters = kmeans.predict(X)

# Print the sentences in each cluster
for cluster_id in range(k):
    cluster_indices = [i for i, c in enumerate(clusters) if c == cluster_id]
    print(f"Group {cluster_id + 1}:")
    print(f"Number of claims: {len(cluster_indices)}")
    print("claims:")
    for idx in cluster_indices:
        print(claims[idx])
    print("\n")

```

LDA:

LDA is a generative statistical model that allows sets of observations to be explained by unobserved groups.
It assumes that each document is a mixture of topics and each word's presence is attributable to one of the document's topics.
By applying LDA, you can identify the underlying topics in your patent claim sentences and assign each sentence to the most relevant topic.

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer for converting text claims into numerical vectors
vectorizer = CountVectorizer()

# Transform the claims into a document-term matrix
X = vectorizer.fit_transform(claims)

# Define the number of topics
num_topics = 3

# Initialize the Latent Dirichlet Allocation (LDA) model with the specified number of topics
lda = LatentDirichletAllocation(n_components=num_topics)

# Fit the LDA model to the document-term matrix
lda.fit(X)

# Assign each claim to its corresponding topic
topics = lda.transform(X).argmax(axis=1)

# Print the claims in each topic along with the topic number
for topic_id in range(num_topics):
    # Get the indices of claims belonging to the current topic
    topic_indices = [i for i, t in enumerate(topics) if t == topic_id]
    
    # Print topic information
    print(f"Group {topic_id + 1}:")
    print(f"Number of claims: {len(topic_indices)}")
    print("Claims:")
    
    # Print each claim in the topic
    for idx in topic_indices:
        print(claims[idx])
    
    print("\n")  # Print a newline to separate topics

```

Hierarchical Clustering:

Hierarchical clustering is another clustering approach that builds a tree of clusters.
It does not require specifying the number of clusters in advance (but it possible), making it more flexible than K-means.
In this method, you start with each sentence as a separate cluster and iteratively merge them based on their similarity until all sentences belong to a single cluster.
The resulting tree can be visualized as a dendrogram, which allows you to inspect the hierarchical structure of the clusters.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.cluster.hierarchy import linkage, fcluster
from collections import Counter

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Transform the claims into TF-IDF vectors
X = vectorizer.fit_transform(claims)

# Compute the linkage matrix using Ward's method
linkage_matrix = linkage(X.toarray(), 'ward')

# You can manually decide the number of clusters (groups)
num_groups = 3  # Number of groups

# Assign clusters using the maximum cluster criterion ('maxclust')
clusters = fcluster(linkage_matrix, num_groups, criterion='maxclust')

# Print the claims in each group along with the group number
for group_id in range(1, num_groups + 1):
    # Get the indices of claims belonging to the current group
    group_indices = [i for i, c in enumerate(clusters) if c == group_id]
    
    # Print group information
    print(f"Group {group_id}:")
    print(f"Number of claims: {len(group_indices)}")
    print("Claims:")
    
    # Print each claim in the group
    for idx in group_indices:
        print(claims[idx])
    
    print("\n")  


```
